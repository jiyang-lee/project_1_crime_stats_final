```python
import pandas as pd
```

```python
# 1. 경찰청 지역, 시간, 요일 데이터 부르기

df_region = pd.read_csv("../data/police_region_fix.csv", encoding="utf-8-sig")
df_time = pd.read_csv("../data/police_time_fix.csv", encoding="utf-8-sig")
df_week = pd.read_csv("../data/police_week_fix.csv", encoding="utf-8-sig")
```

```python
# 범죄중분류가 강제추행인 데이터를 요일별로 시각화 - 정렬: 요일, 시각화: 시계열 사용툴: plotly express
import plotly.express as px
# 강제추행 데이터 필터링
df_chujang = df_week[df_week['범죄중분류'] == '절도']
# 요일 순서 정의
weekday_order = ['월', '화', '수', '목', '금', '토', '일']

# 요일별 발생건수 합계 계산
chujang_weekly = df_chujang.groupby(['요일'])['발생건수'].sum().reindex(weekday_order).reset_index()
# 시계열 그래프 그리기
fig = px.line(chujang_weekly, x='요일', y='발생건수', title='요일별 절도 발생건수')
fig.show()
```
