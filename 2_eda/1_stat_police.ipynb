## 2. 오늘의 집중 주의 범죄 (Spotlight Crime)

단순히 건수가 많은 범죄가 아니라, **평소보다 오늘 유독 비중이 높아지는 범죄**를 콕 집어 제공합니다.  
(예: 평소엔 절도가 1위지만, 오늘은 폭행 비중이 급증하는 경우)

### 📌 분석 로직 (Pandas)

- **전체 데이터 구성비**
  
  `total_share = df.groupby('범죄중분류')['발생건수'].sum() / df['발생건수'].sum()`

- **오늘 요일 구성비**
  
  `today_share = df[df['요일'] == '오늘'].groupby('범죄중분류')['발생건수'].sum() / 오늘의_총_발생건수`

- **상승 폭 계산**
  
  `diff = today_share - total_share`

- **가장 큰 양수값을 가진 범죄 추출**



# 4. 시간대별 통계 - 위험도


````python
df_time
````

````python
# 시간대별 총합 / 시간대별 평균

time_sum = df_time.groupby('시간대')['범죄건수'].sum()
time_avg = time_sum.mean()  
danger_score_time = (time_sum / time_avg) * 100
danger_score_time
````

- 아침 6시부터 9시가 제일 안전
- 밤 9시부터 11시가 제일 위험



````python
daytime_total = df_time[df_time['시간대'].str.contains('09시|12시|15시')]['범죄건수'].sum()

# 2. 전체 대비 낮 시간대 비중 (%) 계산
total_sum = df_time['범죄건수'].sum()
day_ratio = (daytime_total / total_sum) * 100

print(f"낮 시간대 범죄 비중: {day_ratio:.1f}%")
````

````python
"낮에 유독 많이 발생하는 범죄 Top 3는?"
df_time[df_time['시간대'].str.contains('09시|12시|15시')].groupby('범죄중분류')['범죄건수'].sum().sort_values(ascending=False)
````

````python
# 1. 낮 시간대(09-18시)만 필터링
daytime_df = df_time[df_time['시간대'].str.contains('09시|12시|15시')]

# 2. 어떤 범죄가 가장 많은지 순위 계산
daytime_rank = daytime_df.groupby('범죄중분류')['범죄건수'].sum().sort_values(ascending=False)

# 3. 전체 평균과 비교해서 낮에 유독 튀는 범죄 찾기 (배수 계산)
overall_avg = df_time.groupby('범죄중분류')['범죄건수'].sum() / 8 # 8개 시간대 기준
lift = (daytime_rank / overall_avg).sort_values(ascending=False)
````

# 6. 시각화

````python
#범죄중분류가 강도인 데이터만 필터링, 요일에서 loc로
df_week.loc[df_week['범죄중분류'] == '절도범죄']
````

````python
# 범죄중분류가 강제추행인 데이터를 요일별로 시각화 - 정렬: 요일, 시각화: 시계열 사용툴: plotly express
import plotly.express as px
# 강제추행 데이터 필터링
df_chujang = df_week[df_week['범죄중분류'] == '절도범죄']
# 요일 순서 정의
weekday_order = ['월', '화', '수', '목', '금', '토', '일']

# 요일별 발생건수 합계 계산
chujang_weekly = df_chujang.groupby(['요일'])['발생건수'].sum().reindex(weekday_order).reset_index()
# 시계열 그래프 그리기
fig = px.line(chujang_weekly, x='요일', y='발생건수', title='요일별 절도범죄 발생건수')
fig.show()
````

````python
df_region['범죄중분류'].unique()
````

````python
#  절도범죄인 데이터 행 절도로 변경

targets = '절도범죄'
new_label = '절도'

for df in (df_region, df_week):
    df['범죄중분류'] = df['범죄중분류'].astype(str).str.strip()
    df.loc[df['범죄중분류'] == targets, '범죄중분류'] = new_label
````

````python
targets = '마약범죄'
new_label = '마약'

for df in (df_region, df_week, df_time):
    df['범죄중분류'] = df['범죄중분류'].astype(str).str.strip()
    df.loc[df['범죄중분류'] == targets, '범죄중분류'] = new_label
````
