```python
import pandas as pd

# 1. 파일 불러오기
df = pd.read_csv('../data/police_time_fix.csv')

# 2. 새로운 칼럼 '시분_범위' 만들기
# '시'를 ':'로 바꾸고 '분'을 없애는 작업을 전체 문자열에 적용합니다.
df['시간대2'] = df['시간대'].str.replace('시', ':').str.replace('분', '')

# 3. 결과 확인
print(df[['시간대', '시간대2']].head())

# 4. CSV로 저장
df.to_csv('../data/police_time_fix.csv', index=False, encoding='utf-8-sig')
```
